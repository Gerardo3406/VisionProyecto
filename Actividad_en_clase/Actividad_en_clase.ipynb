# Mi primer proyecto de visión computacional
Este primer proyecto es un ejemplo de cómo utilizar OpenCVde maner efectiva

```python
# # Importar librerias
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
bgr = cv2.imread('images.jpg')
print("El tamaño de la imagen es: ", bgr.shape)

# Convertir la imagen de BGR a RGB
rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)

# Mostrar la imagen
plt.imshow(rgb)
```

```python
# Mostrar la imagen en escala de grises
gray = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
#Mostrar las componentes de color de color de la imagen
b, g, r = cv2.split(bgr)
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.imshow(b, cmap='gray')
plt.title('Canal Azul')
plt.subplot(1,3,2)
plt.imshow(g, cmap='gray')
plt.title('Canal Verde')
plt.subplot(1,3,3)
plt.imshow(r, cmap='gray')
plt.title('Canal Rojo')
plt.show()
```

```python
# Mostrar las componentes HLS de la imagen
hls = cv2.cvtColor(bgr, cv2.COLOR_BGR2HLS)
h, l, s = cv2.split(hls)
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.imshow(h, cmap='gray')
plt.title('Canal Hue')
plt.subplot(1,3,2)
plt.imshow(l, cmap='gray')
plt.title('Canal Lightness')
plt.subplot(1,3,3)
plt.imshow(s, cmap='gray')
plt.title('Canal Saturation')
plt.show()
```
